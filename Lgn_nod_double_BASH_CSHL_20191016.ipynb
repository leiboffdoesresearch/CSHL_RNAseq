```bash
# Create new conda environment for RNAseq alignment and counting
# the environment is named 'alignment'
# and contains the 3.6 release of python
# the most up-to-date version of fastp for quality control
#                             samtools for general read handling
#                               hisat2 for spliced alignment
#                          and subread for read counting

conda create -y --name alignment python=3.6 fastp samtools hisat2 subread
```

# Intro to RNAseq  
### and our double mutant analysis  

```bash
# activate your new conda environment
conda activate alignment
```

```bash
#Use fastp to look at your sequence files and verify they are ok
#If Illumina adapter sequences are detected at the end of your read, fastp will remove them
#for single end fastp -i in.R1.fq.gz -o out.R1.fq.gz 

#run fastp in command line
mkdir ./fastp_report   #reports go here
mkdir ./trimmed_reads  #trimmed reads go here

# this is a more complicated command
# here, we use the command 'find' to identify all files with same ending name
# and process them one by one -- 'do'-ing the middle stuff
# until there are no more files and we are 'done'

# in the middle, we are looking at the raw reads with fastp
find ./raw_reads -name "*_L001_R1_001.fastq" | while read f; do
STEM=$(basename "${f}" _L001_R1_001.fastq)
fastp -w 4 -i ./raw_reads/${STEM}_L001_R1_001.fastq \
    -o ./trimmed_reads/${STEM}.fastq.gz \
    -h ./fastp_report/${STEM}.html \
    -j ./fastp_report/${STEM}.json \
    -R "${STEM} fastp report"
done
```

```bash
# save your work
# put fastp results into the datastore
icd 
imkdir -p ./CSHL_Cereal_Genomics_2019/finished_analysis
iput -rKPf -N 4 ./fastp_report
iput -rKPf -N 4 ./trimmed_reads
```

```bash
#check that we have some genome index files

ls -l ./maize_genome
```

```bash
# use HiSAT2 to align your reads to the genome
# HiSAT2 is a fast, splicing aware aligner that can accept cDNA inputs

# lets try running HiSAT2 on just one file
# but we will cancel it after just a few seconds
# use control+c to terminate the file after you have seen its output
# how do we determine the options to use?

hisat2 --min-intronlen 60 --max-intronlen 50000 \
    -p 4 -x ./maize_genome/ZmaysAGPv3.30.hisat2 \
    -U ./trimmed_reads/WT10_S8.fastq.gz
    
# this will output a SAM file directly to the command line
```

```bash
# use HiSAT2 to align your reads to the genome
# HiSAT2 is a fast, splicing aware aligner that can accept cDNA inputs

mkdir ./aligned_reads # make destination folder
mkdir ./HiSAT2_report # just for summary stats

#min intron 60 max intron 50000 from Chen et al. 2018
#running in unstranded mode - input files with -U
#output to command line then pipe to samtools view
#and capture stdout with > to make a BAM file
#capture stderr with 2> to make a report file
find ./trimmed_reads -name "*.fastq.gz" | while read f; do
	STEM=$(basename "${f}" .fastq.gz)
	hisat2 --min-intronlen 60 --max-intronlen 50000 \
	-p 4 -x ./maize_genome/ZmaysAGPv3.30.hisat2 \
	-U ./trimmed_reads/"${STEM}".fastq.gz \
    2> ./HiSAT2_report/"${STEM}"_summary.txt \
    | samtools view -bS - \
    > ./aligned_reads/"${STEM}".bam
done
```

```bash
#check on your alignments and reports
ls -lh ./aligned_reads 
ls -lh ./HiSAT2_report
```

```bash
# what does the report look like?
# we can use cat to lump the text together and output it to the command line

cat ./HiSAT2_report/WT10_S8_summary.txt
```

```bash
#save your work
icd 
icd ./CSHL_Cereal_Genomics_2019/finished_analysis
iput -rKPf -N 4 ./aligned_reads
iput -rKPf -N 4 ./HiSAT2_report
```

```bash
## use featureCounts to get expression values
# count by EXON and
# summarize by GENE
# this is an UNION-EXON approach

#command template
#featureCounts -p -a annotation.gtf -t exon -g gene id
#-o counts.txt mapping results PE.bam

mkdir ./featureCounts

for f in ./aligned_reads/*.bam; do
  STEM=$(basename "${f}" .bam)
  featureCounts -T 4 \
      -a ./maize_genome/Zea_mays.AGPv3.30.repaired.gtf \
      -t exon -g gene_id \
      -o ./featureCounts/${STEM}.counts.txt \
      ./aligned_reads/${STEM}.bam
done
```
